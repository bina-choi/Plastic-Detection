```python
import os
import random
import yaml

import numpy as np
import pandas as pd

from ultralytics import YOLO
import shutil
```

```python
model = YOLO('/content/drive/MyDrive/preprocessed_data_/modeling_result/SGD_0.01_50_64_yolov8n/detect/train5/weights/best.pt')
```

```python
# 테스트 결과 구글 드라이브에 저장

source_path = '/content/runs'

# 목적지 경로
destination_path = os.path.join('/content/drive/MyDrive/preprocessed_data_/modeling_result/final_test')

# 목적지 경로가 없으면 생성
if not os.path.exists(destination_path):
    os.makedirs(destination_path)

# '/content/runs' 경로에 있는 모든 파일과 폴더를 이동
for filename in os.listdir(source_path):
    file_path = os.path.join(source_path, filename)
    shutil.move(file_path, destination_path)
```
